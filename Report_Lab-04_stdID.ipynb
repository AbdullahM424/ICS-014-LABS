- <font color='blue'>Exercise# 2:</font> 
Write a program that reads a temperature and its unit (C or F) from the user, then it converts it to the other unit as shown in the sample runs.
The following is the formula for converting the temperature from Celsius to Fahrenheit.<br>
F = 9/5 C +32 <br>
Note: the input can be lower case or upper case

````python
#code for execise 2
t= float(input("Enter current temperature: "))      # t means temperature, here we will put the value of temperature.
unit= str(input("Enter it's unit C or F: "))   #Here we will put the value of unit.

if unit=="C":  #Here we will put the unit in celsius.
    print("%0.2F"%t,unit,"=",((t*9)/5)+32,"F")
else:
    if unit=="F":  #Here we will put the unit in fahrenheit.
        print("%0.2F"%t,unit,"=",round(((t-32)*5)/9,2),"C")
    else:    #Here if it is not celsius and fahrenheit.
        print("Worng unit")
    
````
